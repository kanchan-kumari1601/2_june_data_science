```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\ASUS\OneDrive\Desktop\Internship 2june\Deep learning\ANN\clean_Credit_RNN.csv')
x=df.drop('approval',axis=1)
y =df['approval']
```

```python
x
```

```python
input_shape = x.shape[1]
```

```python

input_shape
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size=0.2)
```

```python
model =Sequential()
model.add(Dense(units=128, activation='relu', input_shape=(input_shape,)))  ### input layer

# Dropout layer 
model.add(Dropout(0.2))

### hidden layer 
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=32, activation='relu'))
model.add(Dense(units=16, activation='relu'))
model.add(Dense(units=8, activation='relu'))



# output layer 

model.add(Dense(units=1, activation='sigmoid')) 



model.compile(loss = 'binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(x_train , y_train , epochs=10)
```

```python
model.predict(  x_test)
```
