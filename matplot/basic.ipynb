```python
x=np.array([1,2,3,4])
y=np.array([4,5,6,7])
plt.title("x and y data point")
plt.xlabel("x data point")
plt.ylabel("y data point")
plt.grid()
plt.plot(x)
plt.plot(y)
plt.show()
```

```python
x=np.array([1,2,3,4])
y=np.array([4,5,6,7])
plt.title("x and y data point")
plt.xlabel("x data point")
plt.ylabel("y data point")
plt.grid()
plt.plot(x,color='r')
plt.plot(y,color='g')
plt.show()
```

## Scatter plot

```python
x=np.array([1,2,3,4,5])
y=np.array([1,2,4,6,7])
plt.scatter(x,y)
plt.show()
```

```python
x=np.array([1,2,3,4,5])
y=np.array([1,2,4,6,7])
plt.scatter(x,y)


x=np.array([1,6,8,4,5])
y=np.array([1,9,4,1,7])
plt.scatter(x,y)
plt.show()
```

```python
x=np.array([1,2,3,4,5])
y=np.array([1,2,4,6,7])
color=np.array(['red','green','blue','yellow','black'])
plt.scatter(x,y,c=color,alpha=0.5)
plt.show()
```

```python
x=np.array([1,2,3,4,5])
y=np.array([1,2,4,6,7])
color=np.array(['red','green','blue','yellow','black'])
size=np.array([10,20,30,40,50])
plt.scatter(x,y,c=color,alpha=0.5,s=size)
plt.show()
```

```python
x=np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])
c = np.array(['red', 'green', 'blue', 'yellow', 'cyan', 'magenta', 'black', 'orange', 'red', 'green', 'blue', 'yellow', 'cyan'])
plt.scatter(x, y, s = sizes, c = c )
plt.show()
```

```python
x=np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])
c = np.array(['red', 'green', 'blue', 'yellow', 'cyan', 'magenta', 'black', 'orange', 'red', 'green', 'blue', 'yellow', 'cyan'])
plt.scatter(x, y, s = sizes, c = c,cmap='coolwarm' )
plt.colorbar()
plt.show()
```

### pie plot
